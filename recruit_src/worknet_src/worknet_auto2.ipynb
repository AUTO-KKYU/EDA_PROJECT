```python
import pandas as pd
import googlemaps

# Google Maps API 키
gmaps_key = "AIzaSyCbQlfd66P6wwp_Wc3SfKwtfZmKjNWNbzE"
gmaps = googlemaps.Client(key=gmaps_key)

# 위도와 경도를 저장할 빈 리스트 생성
latitudes = []
longitudes = []

# 주소를 위도와 경도로 변환하는 함수 정의
def get_lat_lng(address):
    try:
        geocode_result = gmaps.geocode(address)
        location = geocode_result[0]['geometry']['location']
        lat, lng = location['lat'], location['lng']
        return lat, lng
    except (IndexError, KeyError):
        print(f"Error: Failed to get coordinates for address - {address}")
        return None, None

# DataFrame의 각 행에 대해 위도와 경도를 찾아서 리스트에 저장
for index, row in df.iterrows():
    lat, lng = get_lat_lng(row['기업이름'])
    latitudes.append(lat)
    longitudes.append(lng)

# 새로운 열로 추가
df['위도'] = latitudes
df['경도'] = longitudes
df
```

### 지도 시각화
- 현재까지의 데이터 정보를 바탕으로 folium을 활용하여 각 기업의 위치에 대해 표기하였음
- 네이버랩스,코가로보틱스 주식회사, (주)현대엠시스템즈 => 기업이름이 겹치므로 중복되서 지도에 표기되었음

```python
import numpy as np
from tqdm import tqdm

lat_list = []
lng_list = []

for idx, row in tqdm(df.iterrows(), total = len(df.index)):
    geocode = gmaps.geocode(row["기업위치"])
    
    lat = geocode[0].get("geometry")["location"]["lat"] if len(geocode) > 0 else np.nan
    lng = geocode[0].get("geometry")["location"]["lng"] if len(geocode) > 0 else np.nan
    
    lat_list.append(lat)
    lng_list.append(lng)

len(lat_list), len(lng_list)
```

```python
df["lat"] = lat_list
df["lng"] = lng_list

df 
```

```python
# NaN 값을 특정 값으로 대체
df = df.fillna('-')
df
```

```python
mapping = folium.Map(location = [lat_list[0], lng_list[0]], zoom_start=10)

for idx, row in df.iterrows():
    if row["lat"] == np.nan or row["lng"] == np.nan:
        continue
    else:
        folium.Marker([row["lat"], row["lng"]], popup = row["기업이름"]).add_to(mapping)
mapping
```
