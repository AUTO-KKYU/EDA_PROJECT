```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
import pandas as pd
from requests.exceptions import Timeout

base_url = "https://job.incruit.com/jobdb_list/searchjob.asp?occ2=175&occ2=170&occ2=632&occ2=578&scale=3&rgn2=11&rgn2=18"

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(options=chrome_options)

# 데이터프레임 만들기
columns = ["기업이름", "매출액"]
df = pd.DataFrame(columns=columns)

# 기업이름과 매출액에 대한 정보 추출
def extract_company_info(company_url):
    try:
        response = requests.get(company_url, timeout=10)  
        response.raise_for_status()  
        soup = BeautifulSoup(response.text, "html.parser")

        name_element = soup.select_one('#company_warp > div:nth-child(1) > div > div > div.section_view_company_header_right > div > div > div > ul:nth-child(1) > li:nth-child(1) > span')
        revenue_element = soup.select_one('#company_warp > div:nth-child(1) > div > div > div.section_view_company_header_right > div > div > div > ul:nth-child(3) > li:nth-child(1) > span')

        if revenue_element:
            revenue_text = revenue_element.get_text(strip=True)
            name_text = name_element.get_text(strip=True)

            df.loc[len(df)] = [name_text, revenue_text]
            print("기업이름: ", name_text)
            print("매출액:", revenue_text)
        else:
            print("매출액에 대한 정보가 없음")
    except Timeout:
        print("Timeout", company_url)
    except requests.RequestException as e:
        print("Error:", e)

for page_number in range(1, 9):
    url = f"{base_url}&page={page_number}"
    driver.get(url)

    job_elements = driver.find_elements(By.CSS_SELECTOR, '#JobList_Area > div > div.cBbslist_contenst > ul > li > div.cell_first > div.cl_top > a')

    for job_element in job_elements:
        job_url = job_element.get_attribute("href")
        extract_company_info(job_url)
```
